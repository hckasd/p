```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import os
import shutil
import random
from imutils import paths
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import cv2
from ultralytics import YOLO
```

```
!mkdir -p dataset/images dataset/labels
```

```
classes = [
    "BEANS", "CAKE", "CANDY", "CEREAL", "CHIPS", "CHOCOLATE", "COFFEE", "CORN",
    "FISH", "FLOUR", "HONEY", "JAM", "JUICE", "MILK", "NUTS", "OIL", "PASTA",
    "RICE", "SODA", "SPICES", "SUGAR", "TEA", "TOMATO_SAUCE", "VINEGAR", "WATER"
]

class_map = {name: idx for idx, name in enumerate(classes)}

# Đường dẫn đến dataset
image_dir = "images"
label_dir = "labels"

os.makedirs(label_dir, exist_ok=True)

for class_name in os.listdir(image_dir):
    class_id = class_map[class_name]
    class_path = os.path.join(image_dir, class_name)

    label_class_dir = os.path.join(label_dir, class_name)
    os.makedirs(label_class_dir, exist_ok=True)

    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        img = cv2.imread(img_path)
        h, w, _ = img.shape

        # Tạo file nhãn YOLO
        label_file = os.path.join(label_class_dir, img_name.replace('.png', '.txt'))
        with open(label_file, 'w') as f:
            x_center, y_center, width, height = 0.5, 0.5, 1.0, 1.0  # Tạm thời set toàn ảnh
            f.write(f"{class_id} {x_center} {y_center} {width} {height}\n")
```

```
classes = [
    "BEANS", "CAKE", "CANDY", "CEREAL", "CHIPS", "CHOCOLATE", "COFFEE", "CORN",
    "FISH", "FLOUR", "HONEY", "JAM", "JUICE", "MILK", "NUTS", "OIL", "PASTA",
    "RICE", "SODA", "SPICES", "SUGAR", "TEA", "TOMATO_SAUCE", "VINEGAR", "WATER"
]

train_images = list(paths.list_images('/content/images'))
sample_images = random.sample(train_images, min(len(train_images), 25))

plt.figure(figsize=(12, 12))
for i, img_path in enumerate(sample_images):
  img = plt.imread(img_path)
  plt.subplot(5, 5, i + 1)
  plt.imshow(img)
  plt.title(img_path.split('/')[-2])
  plt.axis('off')

plt.show()
```

```
%%writefile dataset.yaml
path: /content
train: images
val: images
nc: 25
names: ["BEANS", "CAKE", "CANDY", "CEREAL", "CHIPS", "CHOCOLATE", "COFFEE", "CORN",
        "FISH", "FLOUR", "HONEY", "JAM", "JUICE", "MILK", "NUTS", "OIL", "PASTA",
        "RICE", "SODA", "SPICES", "SUGAR", "TEA", "TOMATO_SAUCE", "VINEGAR", "WATER"]
```

```
model = YOLO("yolov10n.pt")
model.train(data="/content/dataset.yaml", epochs=10, imgsz=640, batch=16)
```
